# Notificator

```python
from gtts import gTTS
from IPython.display import Audio

def send_notification(message="Операция завершена"):
    # Создаем объект gTTS с текстом уведомления
    tts = gTTS(text=message, lang='ru')
    # Сохраняем генерированный аудиофайл
    tts.save('notification.mp3')
    # Воспроизводим уведомление
    return Audio("notification.mp3", autoplay=True)
send_notification()
```

# dataset preparation

```python
emnist_train = torchvision.datasets.EMNIST('./', split='letters', download = True, train = True)
mnist_train = torchvision.datasets.EMNIST('./', split='mnist', download = True, train = True)

device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')

emnist_train_data = emnist_train.data
mnist_train_data = mnist_train.data
train_data = torch.cat([emnist_train_data[:24960], mnist_train_data[:12000]], dim=0).float().unsqueeze(1).to(device) / 255

temp_emnist_train_labels = emnist_train.targets[:24960]
emnist_train_labels = []

for i in range(len(temp_emnist_train_labels)):
  if temp_emnist_train_labels[i]==15:
    emnist_train_labels.append(torch.tensor(0))
  elif temp_emnist_train_labels[i]>15:
    emnist_train_labels.append(temp_emnist_train_labels[i]+8)
  else:
    emnist_train_labels.append(temp_emnist_train_labels[i]+9)

mnist_train_labels = mnist_train.targets
train_labels = torch.cat([torch.stack(emnist_train_labels[:24960]), mnist_train_labels[:12000]], dim=0).to(device)

file = open('MTUCI_EMNIST/dataset.pkl', 'rb')
test_dataset = pickle.load(file)
file.close()

test_data = test_dataset['data'].numpy()
test_data = np.flip(test_data, axis = 3)
test_data = np.rot90(test_data, k=1, axes=(2,3))
test_data = 1 - test_data
test_data = torch.from_numpy(test_data).float().to(device)
test_labels = test_dataset['targets'].to(device)
```

# model configuration

```python
class Lenet(torch.nn.Module):
  def __init__(self):
    super(Lenet, self).__init__()
    self.conv1 = torch.nn.Conv2d(in_channels=1, out_channels=3, kernel_size=3, padding=1)
    self.bn1 = torch.nn.BatchNorm2d(3)
    self.act1 = torch.nn.ReLU6()
    self.dropout1 = torch.nn.Dropout(p=0.05)

    self.conv2 = torch.nn.Conv2d(in_channels=3, out_channels=3, kernel_size=3, padding=1)
    self.bn2 = torch.nn.BatchNorm2d(3)
    self.act2 = torch.nn.ReLU6()
    self.dropout2 = torch.nn.Dropout(p=0.05)

    self.conv3 = torch.nn.Conv2d(in_channels=3, out_channels=3, kernel_size=3, padding=1)
    self.bn3 = torch.nn.BatchNorm2d(3)
    self.act3 = torch.nn.ReLU6()
    self.dropout3 = torch.nn.Dropout(p=0.05)

    self.conv4 = torch.nn.Conv2d(in_channels=3, out_channels=6, kernel_size=3, padding=1)
    self.bn4 = torch.nn.BatchNorm2d(6)
    self.act4 = torch.nn.ReLU6()
    self.dropout4 = torch.nn.Dropout(p=0.05)

    self.conv5 = torch.nn.Conv2d(in_channels=6, out_channels=6, kernel_size=3, padding=1)
    self.bn5 = torch.nn.BatchNorm2d(6)
    self.act5 = torch.nn.ReLU6()
    self.dropout5 = torch.nn.Dropout(p=0.05)

    self.conv6 = torch.nn.Conv2d(in_channels=6, out_channels=6, kernel_size=3, padding=1)
    self.bn6 = torch.nn.BatchNorm2d(6)
    self.act6 = torch.nn.ReLU6()
    self.dropout6 = torch.nn.Dropout(p=0.05)

    self.pool1 = torch.nn.AvgPool2d(kernel_size=2, stride=2)
    self.act7 = torch.nn.ReLU6()

    self.conv7 = torch.nn.Conv2d(in_channels=6, out_channels=11, kernel_size=3, padding=0)
    self.bn7 = torch.nn.BatchNorm2d(11)
    self.act8 = torch.nn.ReLU6()
    self.dropout7 = torch.nn.Dropout(p=0.05)

    self.conv8 = torch.nn.Conv2d(in_channels=11, out_channels=11, kernel_size=3, padding=1)
    self.bn8 = torch.nn.BatchNorm2d(11)
    self.act9 = torch.nn.ReLU6()
    self.dropout8 = torch.nn.Dropout(p=0.05)

    self.conv9 = torch.nn.Conv2d(in_channels=11, out_channels=11, kernel_size=3, padding=1)
    self.bn9 = torch.nn.BatchNorm2d(11)
    self.act10 = torch.nn.ReLU6()
    self.dropout9 = torch.nn.Dropout(p=0.05)

    self.conv10 = torch.nn.Conv2d(in_channels=11, out_channels=16, kernel_size=3, padding=0)
    self.bn10 = torch.nn.BatchNorm2d(16)
    self.act11 = torch.nn.ReLU6()
    self.dropout10 = torch.nn.Dropout(p=0.05)

    self.conv11 = torch.nn.Conv2d(in_channels=16, out_channels=16, kernel_size=3, padding=1)
    self.bn11 = torch.nn.BatchNorm2d(16)
    self.act12 = torch.nn.ReLU6()
    self.dropout11 = torch.nn.Dropout(p=0.05)

    self.conv12 = torch.nn.Conv2d(in_channels=16, out_channels=16, kernel_size=3, padding=1)
    self.bn12 = torch.nn.BatchNorm2d(16)
    self.act13 = torch.nn.ReLU6()
    self.dropout12 = torch.nn.Dropout(p=0.05)

    self.pool2 = torch.nn.AvgPool2d(kernel_size=2, stride=2)
    self.act14 = torch.nn.ReLU6()

    self.fc1 = torch.nn.Linear(16*5*5, 120)
    self.act15 = torch.nn.ReLU6()
    self.fc2 = torch.nn.Linear(120, 84)
    self.act16 = torch.nn.ReLU6()
    self.fc3 = torch.nn.Linear(84, 35)

  def forward(self, x):
    x = self.conv1(x)
    x = self.bn1(x)
    x = self.act1(x)
    x = self.dropout1(x)
    skip_conn1 = x

    x = self.conv2(x)
    x = self.bn2(x)
    x = self.act2(x)
    x = self.dropout2(x)

    x = self.conv3(x)
    x = self.bn3(x)
    x = self.act3(x)
    x = self.dropout3(x)
    x = torch.add(x, skip_conn1)

    x = self.conv4(x)
    x = self.bn4(x)
    x = self.act4(x)
    x = self.dropout4(x)
    skip_conn2 = x

    x = self.conv5(x)
    x = self.bn5(x)
    x = self.act5(x)
    x = self.dropout5(x)

    x = self.conv6(x)
    x = self.bn6(x)
    x = self.act6(x)
    x = self.dropout6(x)
    x = torch.add(x, skip_conn2)

    x = self.pool1(x)
    x = self.act7(x)

    x = self.conv7(x)
    x = self.bn7(x)
    x = self.act8(x)
    x = self.dropout7(x)
    skip_conn3 = x

    x = self.conv8(x)
    x = self.bn8(x)
    x = self.act9(x)
    x = self.dropout8(x)

    x = self.conv9(x)
    x = self.bn9(x)
    x = self.act10(x)
    x = self.dropout9(x)
    x = torch.add(x, skip_conn3)

    x = self.conv10(x)
    x = self.bn10(x)
    x = self.act11(x)
    x = self.dropout10(x)
    skip_conn4 = x

    x = self.conv11(x)
    x = self.bn11(x)
    x = self.act12(x)
    x = self.dropout11(x)

    x = self.conv12(x)
    x = self.bn12(x)
    x = self.act13(x)
    x = self.dropout12(x)
    x = torch.add(x, skip_conn4)

    x = self.pool2(x)
    x = self.act14(x)
    x = x.view(x.size(0), x.size(1)*x.size(2)*x.size(3))

    x = self.fc1(x)
    x = self.act15(x)
    x = self.fc2(x)
    x = self.act16(x)
    x = self.fc3(x)
    return x
```

# AdaBound 70 epoch

```python
loss = torch.nn.CrossEntropyLoss()
lenet = Lenet().to(device)
optimizer = torch_optimizer.AdaBound(lenet.parameters(), lr= 1e-3,
    betas= (0.9, 0.999),
    final_lr = 0.1,
    gamma=1e-3,
    eps= 1e-8,
    weight_decay=0,
    amsbound=False,)
test_accuracy_history = []
batch_size = 1120

for epoch in bar(range(200), desc='learning'):
  order = np.random.permutation(len(train_data))

  for start_index in range(0, len(train_data), batch_size):
    optimizer.zero_grad()

    batch_indexes = order[start_index:start_index+batch_size]

    data_batch = train_data[batch_indexes]
    labels_batch= train_labels[batch_indexes]

    preds = lenet.forward(data_batch)

    loss_val = loss(preds, labels_batch)
    loss_val.backward()

    optimizer.step()

  with torch.no_grad():
    test_preds = lenet.forward(test_data)

    accuracy = (test_preds.argmax(dim=1)==test_labels).float().mean().data.cpu()
    test_accuracy_history.append(accuracy)
send_notification()
```

```python
fig = go.Figure(data=go.Scatter(x=[i for i in range(200)], y=test_accuracy_history, mode='lines+markers'))

fig.update_layout(title="AdaBound", xaxis_title="epoch", yaxis_title="accuracy")

fig.show()
```
