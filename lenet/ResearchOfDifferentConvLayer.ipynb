# Confugiration nets

```python
class Lenet1(torch.nn.Module):
  def __init__(self):
    super(Lenet1, self).__init__()
    self.conv1 = torch.nn.Conv2d(in_channels = 1, out_channels = 6, kernel_size=5, padding = 2)
    self.act1 = torch.nn.ReLU()
    self.pool1 = torch.nn.AvgPool2d(kernel_size=3, stride = 2, padding = 1)
    self.act2 = torch.nn.ReLU()
    self.conv2 = torch.nn.Conv2d(in_channels=6, out_channels=16, kernel_size=5, padding=0)
    self.act3 = torch.nn.ReLU()
    self.pool2 = torch.nn.AvgPool2d(kernel_size=2, stride = 2)
    self.act4 = torch.nn.ReLU()
    self.fc1 = torch.nn.Linear(5*5*16, 120)
    self.act5 = torch.nn.ReLU()
    self.fc2 = torch.nn.Linear(120, 84)
    self.act6 = torch.nn.ReLU()
    self.fc3 = torch.nn.Linear(84, 35)

  def forward(self, x):
    x = self.conv1(x)
    x = self.act1(x)
    x = self.pool1(x)
    x = self.act2(x)
    x = self.conv2(x)
    x = self.act3(x)
    x = self.pool2(x)
    x = self.act4(x)
    x = x.view(x.size(0), x.size(1)*x.size(2)*x.size(3))
    x = self.fc1(x)
    x = self.act5(x)
    x = self.fc2(x)
    x = self.act6(x)
    x = self.fc3(x)
    return x
```

```python
class Lenet2(torch.nn.Module):
  def __init__(self):
    super(Lenet2, self).__init__()
    self.conv1 = torch.nn.Conv2d(in_channels=1, out_channels=4, kernel_size=3, padding=1)
    self.act1 = torch.nn.ReLU()
    self.conv2 = torch.nn.Conv2d(in_channels=4, out_channels=6, kernel_size=3, padding=1)
    self.act2 = torch.nn.ReLU()
    self.pool1 = torch.nn.AvgPool2d(kernel_size=2, stride=2)
    self.act3 = torch.nn.ReLU()
    self.conv3 = torch.nn.Conv2d(in_channels=6, out_channels=12, kernel_size=3, padding=0)
    self.act4 = torch.nn.ReLU()
    self.conv4 = torch.nn.Conv2d(in_channels=12, out_channels=16, kernel_size=3, padding=0)
    self.act5 = torch.nn.ReLU()
    self.pool2 = torch.nn.AvgPool2d(kernel_size=2, stride=2)
    self.act6 = torch.nn.ReLU()
    self.fc1 = torch.nn.Linear(5*5*16, 120)
    self.act7 = torch.nn.ReLU()
    self.fc2 = torch.nn.Linear(120, 84)
    self.act8 = torch.nn.ReLU()
    self.fc3 = torch.nn.Linear(84, 35)

  def forward(self, x):
    x = self.conv1(x)
    x = self.act1(x)
    x = self.conv2(x)
    x = self.act2(x)
    x = self.pool1(x)
    x = self.act3(x)
    x = self.conv3(x)
    x = self.act4(x)
    x = self.conv4(x)
    x = self.act5(x)
    x = self.pool2(x)
    x = self.act6(x)
    x = x.view(x.size(0), x.size(1)*x.size(2)*x.size(3))
    x = self.fc1(x)
    x = self.act7(x)
    x = self.fc2(x)
    x = self.act8(x)
    x = self.fc3(x)
    return x
```

```python
class Lenet3(torch.nn.Module):
  def __init__(self):
    super(Lenet3, self).__init__()
    self.conv1 = torch.nn.Conv2d(in_channels=1, out_channels=4, kernel_size=3, padding=1)
    self.act1 = torch.nn.ReLU()
    self.conv2 = torch.nn.Conv2d(in_channels=4, out_channels=4, kernel_size=3, padding=1)
    self.act2 = torch.nn.ReLU()
    self.conv3 = torch.nn.Conv2d(in_channels=4, out_channels=6, kernel_size=3, padding=1)
    self.act3 = torch.nn.ReLU()
    self.conv4 = torch.nn.Conv2d(in_channels=6, out_channels=6, kernel_size=3, padding=1)
    self.act4 = torch.nn.ReLU()
    self.pool1 = torch.nn.AvgPool2d(kernel_size=2, stride=2)
    self.act5 = torch.nn.ReLU()
    self.conv5 = torch.nn.Conv2d(in_channels=6, out_channels=12, kernel_size=3, padding=0)
    self.act6 = torch.nn.ReLU()
    self.conv6 = torch.nn.Conv2d(in_channels=12, out_channels=12, kernel_size=3, padding=1)
    self.act7 = torch.nn.ReLU()
    self.conv7 = torch.nn.Conv2d(in_channels=12, out_channels=16, kernel_size=3, padding=0)
    self.act8 = torch.nn.ReLU()
    self.conv8 = torch.nn.Conv2d(in_channels=16, out_channels=16, kernel_size=3, padding=1)
    self.act9 = torch.nn.ReLU()
    self.pool2 = torch.nn.AvgPool2d(kernel_size=2, stride=2)
    self.act10 = torch.nn.ReLU()
    self.fc1 = torch.nn.Linear(16*5*5, 120)
    self.act11 = torch.nn.ReLU()
    self.fc2 = torch.nn.Linear(120, 84)
    self.act12 = torch.nn.ReLU()
    self.fc3 = torch.nn.Linear(84, 35)

  def forward(self, x):
    x = self.conv1(x)
    x = self.act1(x)
    x = self.conv2(x)
    x = self.act2(x)
    x = self.conv3(x)
    x = self.act3(x)
    x = self.conv4(x)
    x = self.act4(x)
    x = self.pool1(x)
    x = self.act5(x)
    x = self.conv5(x)
    x = self.act6(x)
    x = self.conv6(x)
    x = self.act7(x)
    x = self.conv7(x)
    x = self.act8(x)
    x = self.conv8(x)
    x = self.act9(x)
    x = self.pool2(x)
    x = self.act10(x)
    x = x.view(x.size(0), x.size(1)*x.size(2)*x.size(3))
    x = self.fc1(x)
    x = self.act11(x)
    x = self.fc2(x)
    x = self.act12(x)
    x = self.fc3(x)
    return x
```

```python
class Lenet4(torch.nn.Module):
  def __init__(self):
    super(Lenet4, self).__init__()
    self.conv1 = torch.nn.Conv2d(in_channels=1, out_channels=4, kernel_size=3, padding=2)
    self.act1 = torch.nn.ReLU()
    self.conv2 = torch.nn.Conv2d(in_channels=4, out_channels=4, kernel_size=3, padding=1)
    self.act2 = torch.nn.ReLU()
    self.conv3 = torch.nn.Conv2d(in_channels=4, out_channels=4, kernel_size=3, padding=1)
    self.act3 = torch.nn.ReLU()
    self.conv4 = torch.nn.Conv2d(in_channels=4, out_channels=6, kernel_size=3, padding=1)
    self.act4 = torch.nn.ReLU()
    self.conv5 = torch.nn.Conv2d(in_channels=6, out_channels=6, kernel_size=3, padding=1)
    self.act5 = torch.nn.ReLU()
    self.conv6 = torch.nn.Conv2d(in_channels=6, out_channels=6, kernel_size=3, padding=1)
    self.act6 = torch.nn.ReLU()
    self.pool1 = torch.nn.AvgPool2d(kernel_size=2, stride=2)
    self.act7 = torch.nn.ReLU()
    self.conv7 = torch.nn.Conv2d(in_channels=6, out_channels=12, kernel_size=3, padding=0)
    self.act8 = torch.nn.ReLU()
    self.conv8 = torch.nn.Conv2d(in_channels=12, out_channels=12, kernel_size=3, padding=1)
    self.act9 = torch.nn.ReLU()
    self.conv9 = torch.nn.Conv2d(in_channels=12, out_channels=12, kernel_size=3, padding=1)
    self.act10 = torch.nn.ReLU()
    self.conv10 = torch.nn.Conv2d(in_channels=12, out_channels=16, kernel_size=3, padding=0)
    self.act11 = torch.nn.ReLU()
    self.conv11 = torch.nn.Conv2d(in_channels=16, out_channels=16, kernel_size=3, padding=1)
    self.act12 = torch.nn.ReLU()
    self.conv12 = torch.nn.Conv2d(in_channels=16, out_channels=16, kernel_size=3, padding=1)
    self.act13 = torch.nn.ReLU()
    self.pool2 = torch.nn.AvgPool2d(kernel_size=2, stride=2)
    self.act14 = torch.nn.ReLU()
    self.fc1 = torch.nn.Linear(16*5*5, 120)
    self.act15 = torch.nn.ReLU()
    self.fc2 = torch.nn.Linear(120, 84)
    self.act16 = torch.nn.ReLU()
    self.fc3 = torch.nn.Linear(84, 35)

  def forward(self, x):
    x = self.conv1(x)
    x = self.act1(x)
    x = self.conv2(x)
    x = self.act2(x)
    x = self.conv3(x)
    x = self.act3(x)
    x = self.conv4(x)
    x = self.act4(x)
    x = self.conv5(x)
    x = self.act5(x)
    x = self.conv6(x)
    x = self.act6(x)
    x = self.pool1(x)
    x = self.act7(x)
    x = self.conv7(x)
    x = self.act8(x)
    x = self.conv8(x)
    x = self.act9(x)
    x = self.conv9(x)
    x = self.act10(x)
    x = self.conv10(x)
    x = self.act11(x)
    x = self.conv11(x)
    x = self.act12(x)
    x = self.conv12(x)
    x = self.act13(x)
    x = self.pool2(x)
    x = self.act14(x)
    x = x.view(x.size(0), x.size(1)*x.size(2)*x.size(3))
    x = self.fc1(x)
    x = self.act15(x)
    x = self.fc2(x)
    x = self.act16(x)
    x = self.fc3(x)
    return x
```

```python
class Lenet1(torch.nn.Module):
  def __init__(self):
    super(Lenet1, self).__init__()
    self.conv1 = torch.nn.Conv2d(in_channels=1, out_channels=4, kernel_size=3, padding=1)
    self.act1 = torch.nn.ReLU()
    self.conv2 = torch.nn.Conv2d(in_channels=4, out_channels=4, kernel_size=3, padding=1)
    self.act2 = torch.nn.ReLU()
    self.conv3 = torch.nn.Conv2d(in_channels=4, out_channels=6, kernel_size=3, padding=1)
    self.act3 = torch.nn.ReLU()
    self.conv4 = torch.nn.Conv2d(in_channels=6, out_channels=6, kernel_size=3, padding=1)
    self.act4 = torch.nn.ReLU()
    self.pool1 = torch.nn.AvgPool2d(kernel_size=2, stride=2)
    self.act5 = torch.nn.ReLU()
    self.conv5 = torch.nn.Conv2d(in_channels=6, out_channels=12, kernel_size=3, padding=0)
    self.act6 = torch.nn.ReLU()
    self.conv6 = torch.nn.Conv2d(in_channels=12, out_channels=12, kernel_size=3, padding=1)
    self.act7 = torch.nn.ReLU()
    self.conv7 = torch.nn.Conv2d(in_channels=12, out_channels=16, kernel_size=3, padding=0)
    self.act8 = torch.nn.ReLU()
    self.conv8 = torch.nn.Conv2d(in_channels=16, out_channels=16, kernel_size=3, padding=1)
    self.act9 = torch.nn.ReLU()
    self.pool2 = torch.nn.AvgPool2d(kernel_size=2, stride=2)
    self.act10 = torch.nn.ReLU()
    self.fc1 = torch.nn.Linear(16*5*5, 120)
    self.act11 = torch.nn.ReLU()
    self.fc2 = torch.nn.Linear(120, 84)
    self.act12 = torch.nn.ReLU()
    self.fc3 = torch.nn.Linear(84, 35)

  def forward(self, x):
    x = self.conv1(x)
    x = self.act1(x)
    x = self.conv2(x)
    x = self.act2(x)
    x = self.conv3(x)
    x = self.act3(x)
    x = self.conv4(x)
    x = self.act4(x)
    x = self.pool1(x)
    x = self.act5(x)
    x = self.conv5(x)
    x = self.act6(x)
    x = self.conv6(x)
    x = self.act7(x)
    x = self.conv7(x)
    x = self.act8(x)
    x = self.conv8(x)
    x = self.act9(x)
    x = self.pool2(x)
    x = self.act10(x)
    x = x.view(x.size(0), x.size(1)*x.size(2)*x.size(3))
    x = self.fc1(x)
    x = self.act11(x)
    x = self.fc2(x)
    x = self.act12(x)
    x = self.fc3(x)
    return x
```
